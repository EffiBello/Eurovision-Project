```python
def move_page_euroWorld(currYear):
    print(f'currYear is {currYear}')
    if currYear>=2016:
        driver.find_element_by_xpath('//*[@id="page"]/div[2]/main/div[1]/div[1]/div[2]/a[2]').send_keys("\n")
        if currYear==2020:
            time.sleep(2)
            driver.find_element_by_xpath('//*[@id="page"]/div[2]/main/div/div[1]/div[2]/a[2]').click()
    else:
        driver.find_element_by_xpath('//*[@id="page"]/div[2]/main/div/div[1]/div[2]/a[2]').send_keys("\n")
```

```python
#from 2008 we have two semi final 
#if flag=0 first semi final else second semi final
flag=0

#lists to crawl from eurovisionworld generalBoard
currYear=2004
place=[] #Rank of the competitors
country=[] #Competitors countries list
song=[] #Competitors songs list
artist=[] #Competitors artists list
years=[] #List of the years
point=[] #How many point the recived in total
running=[] #Running order
contestLoc=[] #The place the contest took place
Sfinal=[] #Final/Semi final

#if the suffix is 2 it takes data from the second semi final

#Contains the amount of points each country gave to her competitors for the current year
vote_array=[]
vote_array2=[]

#Contains all the votes every country had for the cuurent year
vote_dict={}
vote_dict2={}

#Contains all the data about the scorebored for every year
vote_dict_by_year={}
vote_dict_by_year2={}

#Countries that gave the points
to_countries=[]
to_countries2=[]

#Countries that recives the points
from_countries=[]
from_countries2=[]
```

```python
driver.get("https://eurovisionworld.com/eurovision/2004")

while currYear<=2021:
    #taking the location of the eurovision
    location=driver.find_element_by_xpath('//*[@id="page"]/div[2]/main/div/div[1]/div[1]/span/span/a').text
    location=re.findall('([a-zA-Z ]*)\d*.*',location)[0]

    #checking how many semi finals the contest had
    if currYear<2008:
        flag=0#one semi final
    else:
        flag=1#two semi final
    
    
    while(flag<3):
        #second semi
        if flag==2: 
            driver.find_element_by_xpath('//*[@id="page"]/div[2]/main/div/div[3]/div[1]/a[3]').send_keys('\n')#press the semi final2 button
        #first semi
        else:
            driver.find_element_by_xpath('//*[@id="page"]/div[2]/main/div/div[4]/div[1]/a[2]').send_keys('\n')#press the semi final 1 button
       
        time.sleep(2)
        #taking data from the general board
        driver.find_element_by_xpath('//*[@id="voting_table"]/table/thead/tr/th[2]').click()#the firstpart of the page
        general_Table =driver.find_element_by_xpath('//*[@id="voting_table"]/table/tbody')

    #from the table on EurovisionWorld General table

        for row in general_Table.find_elements_by_css_selector('tr'):
            place.append(row.find_elements_by_tag_name('td')[0].text)
            country.append(row.find_elements_by_tag_name('td')[1].text)

            #spliting the song and artist from the same cell:
            song.append(row.find_elements_by_tag_name('td')[2].text.split('\n')[0])
            artist.append(row.find_elements_by_tag_name('td')[2].text.split('\n')[1])
           
            point.append(row.find_elements_by_tag_name('td')[3].text)
            if currYear>=2016:
                running.append(row.find_elements_by_tag_name('td')[6].text)
            else:
                running.append(row.find_elements_by_tag_name('td')[4].text)
            years.append(currYear)
            Sfinal.append('Semi Final')
            contestLoc.append(location)

        #extracting information from the score board
        if currYear>=2016:
            driver.find_element_by_xpath('//*[@id="scoreboard"]/div[1]/button[3]').send_keys("\n")
            scoring_Table=driver.find_element_by_xpath('//*[@id="scoreboard"]/div[2]/div[1]/table/tbody')
            table_head=driver.find_element_by_xpath('//*[@id="scoreboard"]/div[2]/div[1]/table/thead')
        else:
            scoring_Table=driver.find_element_by_xpath('//*[@id="scoreboard"]/div/div[1]/table/tbody')
            table_head=driver.find_element_by_xpath('//*[@id="scoreboard"]/div/div[1]/table/thead/tr')
        to_countries=[]
        
        i=1
        from_countries=[]
        #extracting the countries that gave the points
        for country_name in table_head.find_elements_by_css_selector('td'):
            if i>1:
                c_name=country_name.get_attribute('data-from')
                if flag==2: #second semi
                    from_countries2.append(c_name)
                else: #first semi
                    from_countries.append(c_name)
            i+=1
        
        #extracting the points each country gave#
        
        i=0

        #second semi
        if flag==2: 
            for name in from_countries2:
                for row in scoring_Table.find_elements_by_css_selector('tr'):
                    vote_array2.append(row.find_elements_by_tag_name('td')[4+i].text)
                vote_dict2[country_dict_Sc[name.upper()].title()]=vote_array2
                vote_array2=[]
                i+=1

            for row in scoring_Table.find_elements_by_css_selector('tr'):
                to_countries2.append((row.find_elements_by_css_selector('td')[2].text).title())

        #first semi
        else: 
            for name in from_countries:
                for row in scoring_Table.find_elements_by_css_selector('tr'):
                    vote_array.append(row.find_elements_by_tag_name('td')[4+i].text)
                vote_dict[country_dict_Sc[name.upper()].title()]=vote_array
                vote_array=[]
                i+=1
            
            for row in scoring_Table.find_elements_by_css_selector('tr'):
                to_countries.append((row.find_elements_by_css_selector('td')[2].text).title())

        if flag ==1: #first semi final if there are two
             vote_dict_by_year[currYear]=[to_countries,vote_dict]
    
        flag+=1
        if flag==1:
            break

    if flag==1:#if there is only one semi it will enter here
        vote_dict_by_year[currYear]=[to_countries,vote_dict]
    else: # if there are two semi the second will enter here
        vote_dict_by_year2[currYear]=[to_countries2,vote_dict2]

    #reset
    to_countries=[]
    from_countries=[]
    to_countries=[]
    vote_dict={}
    to_countries2=[]
    from_countries2=[]
    vote_dict2={}
    currYear+=1
    move_page_euroWorld(currYear)
    
    if currYear==2020:
        currYear+=1
```

```python
allYearVotes = {} #AllYearVote = {2004:{'be' = {},'fr' = {} }, '1976:{}}
allVotesDict={} #allVoteDict = {'be': {},'fr':{} }
# index_name_dict={}

#running from 2004-2021
for year in range(2004,2021):
    if year==2020:
      year+=1
    ScoreBoardDF=pd.DataFrame(vote_dict_by_year[year][1])
    ScoreBoardDF.index=vote_dict_by_year[year][0]
    for col in ScoreBoardDF:
      ScoreBoardDF[col]= pd.to_numeric(ScoreBoardDF[col], downcast="float")
    
    for col in ScoreBoardDF.columns:
        givenPoints = {}
        arr = ScoreBoardDF[col].values
    
        for i in range(0,len(arr)):
          if(arr[i] > 0): #So we don't get NaN's
              givenPoints[arr[i]] = ScoreBoardDF.index[i]
   
        allVotesDict[col] = OrderedDict(sorted(givenPoints.items())) #Sort the allVoteDict by KEYS!
        givenPoints={}
    
    index_name_dict={}
    allYearVotes[year] = allVotesDict
    allVotesDict= {}

#first year dataframe    
FirstYearDF = pd.DataFrame(allYearVotes[2004])
FirstYearDF=FirstYearDF.T
FirstYearDF['Year']=2004
FirstYearDF
for row in FirstYearDF.index:
  if not(row in vote_dict_by_year[2004][0]):
    FirstYearDF.drop(labels=row,axis=0,inplace=True)
  FirstYearDF.sort_index(inplace=True)
FirstYearDF
```

```python
driver.get("https://eurovisionworld.com/eurovision/2016")
while currYear<=2021:
    flag=1
        
    while(flag<3):
        #second semi
        if flag==2: 
            driver.find_element_by_xpath('//*[@id="page"]/div[2]/main/div[1]/div[3]/div[1]/a[3]').send_keys('\n')  
        #first semi
        else:
            driver.find_element_by_xpath('//*[@id="page"]/div[2]/main/div[1]/div[4]/div[1]/a[2]').send_keys('\n') 
        #getting Crowd's votes
        time.sleep(2)
        #taking data from the scoreboard
        driver.find_element_by_xpath('//*[@id="scoreboard"]/div[1]/button[2]').send_keys('\n') 
        scoring_Table=driver.find_element_by_xpath('//*[@id="scoreboard"]/div[2]/div[1]/table/tbody')
        table_head=driver.find_element_by_xpath('//*[@id="scoreboard"]/div[2]/div[1]/table/thead')
        to_countries=[]
        
        i=1
        from_countries=[]
        #extracting the countries that gave the points
        for country_name in table_head.find_elements_by_css_selector('td'):
            if i>1:
                c_name=country_name.get_attribute('data-from')
                if flag==2:
                    from_countries1.append(c_name)
                else:
                    from_countries.append(c_name)
            i+=1

         #extracting the points each country gave#
        i=0

        #second semi
        if flag==2:
            for name in from_countries1:
                for row in scoring_Table.find_elements_by_css_selector('tr'):
                    vote_array1.append(row.find_elements_by_tag_name('td')[4+i].text)
                vote_dict1[country_dict_Sc[name.upper()].title()]=vote_array1
                vote_array1=[]
                i+=1

            for row in scoring_Table.find_elements_by_css_selector('tr'):
                to_countries1.append((row.find_elements_by_css_selector('td')[2].text).title())
        #first semi
        else:
            for name in from_countries:
                for row in scoring_Table.find_elements_by_css_selector('tr'):
                    vote_array.append(row.find_elements_by_tag_name('td')[4+i].text)
                vote_dict[country_dict_Sc[name.upper()].title()]=vote_array
                vote_array=[]
                i+=1
            
            for row in scoring_Table.find_elements_by_css_selector('tr'):
                to_countries.append((row.find_elements_by_css_selector('td')[2].text).title())

        if flag ==1: #first semi final if there are two
             vote_dict_by_year[currYear]=[to_countries,vote_dict]
    
        flag+=1
        if flag==1:
            break

    if flag==1: #if there is only one semi it will enter here
        vote_dict_by_year[currYear]=[to_countries,vote_dict]
    else: # if there are two semi the second will enter here
        vote_dict_by_year1[currYear]=[to_countries1,vote_dict1]

    #reset
    to_countries=[]
    from_countries=[]
    to_countries=[]
    vote_dict={}
    from_countries1=[]
    to_countries1=[]
    vote_dict1={}
    currYear+=1
    move_page_euroWorld(currYear)
    
    if currYear==2020:
        currYear+=1
```
