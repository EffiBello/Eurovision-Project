```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
# import url
# turn to soup object

URL = "https://en.wikipedia.org/wiki/List_of_Eurovision_Song_Contest_entries_(1956%E2%80%932003)"
page = requests.get(URL)

soup = BeautifulSoup(page.content , "html.parser")
```

```python
result = soup.find(id = "mw-content-text") #table location

tables = result.find_all("table", class_ = "wikitable plainrowheaders") #iter trhough tables

```

```python
for i,table in enumerate(tables): #runs on each table
    # year of the table: 1975->i=19
    year = [int(s) for s in table.find("a").text.split() if s.isdigit()]
    year = int("".join([str(integer) for integer in year]))
    
    if i<19: continue
    print(year, i)

    df=pd.read_html(str(table)) #magic fuck you shler
    # convert list to dataframe
    df=pd.DataFrame(df[0])
    
    #fix placing(string->float, removed special characters):
    df["Placing"]= df["Placing"].apply(lambda x: x.split()[0]) 
    df['Placing'] = pd.to_numeric(df["Placing"], errors= 'coerce')
    df = df.sort_values("Country",ascending=True) #sort each year by the counter Alphabet

    df['Year'] = year #add year to the dataframe
    #table Merging into data frame
    if i==19:
        df_main = df.copy()
    if i>19:
        df_main = df_main.append(df)
    

    
    if year == 2003: break
    

df_main.tail(50)
    
    
    
```

```python
#extract how many songwriters for each song:
counter = 1 #default:1
songwriters=[]
for row in df_main['Songwriter(s)']: #big brain idea
    for i,letter in enumerate(row):
        if i>1:
            if letter.isupper() and row[i-1].islower() : counter += 1
        
    songwriters.append(counter)
    counter=1

print(songwriters) #check
df_main['Number of Songwriter(s)'] = songwriters #add to dataframe
df_main
```

```python
#extract how many languages for each song:
counter = 1 #default :1
languages=[]
for row in df_main['Language']: #another big brain idea
    for letter in  row:
        if letter ==',' : counter += 1
        
    languages.append(counter)
    counter=1

print(languages)        
df_main['Number of Languages'] = languages #add to data frame
df_main
```

```python
#using only relevant columns
final_df = df_main[['R/O','Placing', 'Language', 'Songwriter(s)', 'Number of Languages','Number of Songwriter(s)', 'Year']].copy()
final_df
```

```python
#Bool, Song in English? 1:0
english=[]
for row in final_df["Language"]:
    # print(row, type(row))
    if 'English' in row : english.append(1)
    else : english.append(0)

print(english, len(english))

final_df["English"] = english

final_df
```

```python
#export Df to CSV file
final_df.to_csv('1975-2003.csv', index= False)
```
