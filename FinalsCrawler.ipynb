```python
#lists to crawl from eurovisionworld generalBoard

currYear=1975 #Current year
place=[] #Rank of the competitors
country=[] #Competitors countries list
song=[] #Competitors songs list
artist=[] #Competitors artists list
years=[] #List of the years
point=[] #How many point the recived in total  
running=[] #Running order
contestLoc=[] #The place the contest took place
final=[] #Final/Semi final

vote_array=[] #Contains the amount of points each country gave to her competitors for the current year
vote_dict={} #Contains all the votes every country had for the cuurent year
vote_dict_by_year={} #Contains all the data about the scorebored for every year

to_countries=[] #Countries that gave the points
from_countries=[] #Countries that recives the points
```

in this part we will crawl from eurovisionworld in order to recive the following: place, country,song,artist,year,point,running


```python
driver.get("https://eurovisionworld.com/eurovision/1975") #entering the site

while currYear<=2021:
    #taking the location of the eurovision
    location=driver.find_element_by_xpath('//*[@id="page"]/div[2]/main/div/div[1]/div[1]/span/span/a').text
    location=re.findall('([a-zA-Z ]*)\d*.*',location)[0]

    #taking data from the general board
    driver.find_element_by_xpath('//*[@id="voting_table"]/table/thead/tr/th[2]').click()
    general_Table =driver.find_element_by_xpath('//*[@id="voting_table"]/table/tbody')
    
    #from the table on EurovisionWorld 
    for row in general_Table.find_elements_by_css_selector('tr'):
        place.append(row.find_elements_by_tag_name('td')[0].text)
        country.append(row.find_elements_by_tag_name('td')[1].text)

        #spliting the song and artist from the same cell:
        song.append(row.find_elements_by_tag_name('td')[2].text.split('\n')[0])
        artist.append(row.find_elements_by_tag_name('td')[2].text.split('\n')[1])

        point.append(row.find_elements_by_tag_name('td')[3].text)#total score for the country in the final
        if currYear>=2016:#from 2016 the table has an extra feature so we need some changes
            running.append(row.find_elements_by_tag_name('td')[6].text)
        else:
            running.append(row.find_elements_by_tag_name('td')[4].text)
        years.append(currYear)
        final.append('Final')
        contestLoc.append(location)

    #extracting information from the score bored
    if currYear>=2016:
        driver.find_element_by_xpath('//*[@id="scoreboard"]/div[1]/button[3]').send_keys("\n")
        scoring_Table=driver.find_element_by_xpath('//*[@id="scoreboard"]/div[2]/div[1]/table/tbody')
        table_head=driver.find_element_by_xpath('//*[@id="scoreboard"]/div[2]/div[1]/table/thead')
    else:
        scoring_Table=driver.find_element_by_xpath('//*[@id="scoreboard"]/div/div[1]/table/tbody')
        table_head=driver.find_element_by_xpath('//*[@id="scoreboard"]/div/div[1]/table/thead/tr')
    
    to_countries=[]
    i=1 
    from_countries=[]
    #extracting the countries that gave the points
    for country_name in table_head.find_elements_by_css_selector('td'):
        if i>1:
            c_name=country_name.get_attribute('data-from')
            from_countries.append(c_name)
        i+=1

    #extracting the points each country gave
    i=0
    for name in from_countries:
        for row in scoring_Table.find_elements_by_css_selector('tr'):
            vote_array.append(row.find_elements_by_tag_name('td')[4+i].text)# extracting the amount of points each country gave to other countries 
        vote_dict[country_dict_Sc[name.upper()].title()]=vote_array #orginizing the data in a dictionary "country name: [points]"
        vote_array=[]
        i+=1
    
    #extracting the country's that received the points
    for row in scoring_Table.find_elements_by_css_selector('tr'):
         to_countries.append((row.find_elements_by_css_selector('td')[2].text).title())
    
    vote_dict_by_year[currYear]=[to_countries,vote_dict] # merging the data and creating one dictionary that holds to_countries,vote_dict

    #reset 
    to_countries=[]
    from_countries=[]
    vote_dict={}
    currYear+=1

    move_page_euroWorld(currYear)
    
    if currYear==2020:
        currYear+=1
        
```

```python
allYearVotes = {} #AllYearVote = {1975:{'be' = {},'fr' = {} }, '1976:{}}
allVotesDict={} #allVoteDict = {'be': {},'fr':{} }
# index_name_dict={}

#running from 1975-2021
for year in range(1975,2021):
    if year==2020:
      year+=1
    ScoreBoredDF=pd.DataFrame(vote_dict_by_year[year][1])
    ScoreBoredDF.index=vote_dict_by_year[year][0]
    for col in ScoreBoredDF:
      ScoreBoredDF[col]= pd.to_numeric(ScoreBoredDF[col], downcast="float")
    
    for col in ScoreBoredDF.columns:
        givenPoints = {}
        arr = ScoreBoredDF[col].values
    
        for i in range(0,len(arr)):
          if(arr[i] > 0): #So we don't get NaN's
              givenPoints[arr[i]] = ScoreBoredDF.index[i]
   
        allVotesDict[col] = OrderedDict(sorted(givenPoints.items())) #Sort the allVoteDict by KEYS!
        givenPoints={}
    
    # index_name_dict={}
    allYearVotes[year] = allVotesDict
    allVotesDict= {}

#creating the ScoreBored for the first Year
FirstYearDF = pd.DataFrame(allYearVotes[1975])
FirstYearDF=FirstYearDF.T
FirstYearDF['Year']=1975
for row in FirstYearDF.index:
  if not(row in vote_dict_by_year[1975][0]):
    FirstYearDF.drop(labels=row,axis=0,inplace=True)

FirstYearDF
```

```python
#creating the dataFrame of the General Board 
dict={'Country':country,'Song':song,'Artist':artist,'F/SF':final,'Year':years,'Points':point,'R/O':running,'Place':place,'Contest_Location':contestLoc }
GeneralBoardDF=pd.DataFrame(dict)

GeneralBoardDF["Points"] = pd.to_numeric(GeneralBoardDF["Points"], downcast="float")

#merging 1975 scorebored with 1975 generalboard
mainDF=GeneralBoardDF[GeneralBoardDF['Year']==1975]
for col in FirstYearDF:
    mainDF[f'Jury_{col}']=FirstYearDF[col].values

#merging the rest of the years
for year in range(1976,2021):
    if year==2020:
        year+=1

    ScoreBoardTempDF = pd.DataFrame(allYearVotes[year])
    ScoreBoardTempDF=ScoreBoardTempDF.T
    ScoreBoardTempDF['Year']=year
    
    #droping a country if not in the final
    for row in ScoreBoardTempDF.index:
      if not(row in vote_dict_by_year[year][0]):
        ScoreBoardTempDF.drop(labels=row,axis=0,inplace=True)

    GeneralBoardTempDF=GeneralBoardDF[GeneralBoardDF['Year']==year]
    for col in ScoreBoardTempDF[0:-1]:
        GeneralBoardTempDF[f'Jury_{col}']=ScoreBoardTempDF[col].values

    mainDF=pd.concat([mainDF,GeneralBoardTempDF])
    
mainDF=mainDF.drop(labels='Jury_Year',axis=1)
mainDF
```

creating the dataset

```python
mainDF.to_csv('Final_Jury.csv',index=False)
```
