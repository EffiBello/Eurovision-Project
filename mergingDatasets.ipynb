```python
#Final_Crowd

Final_Crowd=pd.DataFrame(pd.read_csv('Final_Crowd.csv'))
Final_Jury=pd.DataFrame(pd.read_csv('Final_Jury.csv'))
SF_Jury=pd.DataFrame(pd.read_csv('Semi_Final_Jury.csv'))
SF_Crowd=pd.DataFrame(pd.read_csv('Semi_Final_Crowd.csv'))
```

Creating Semi Final(SF) Data frame

```python
SF=SF_Jury[SF_Jury['Year']==2004]
SF
subCrowd=SF_Crowd[SF_Crowd['Year']==2004]
for col in subCrowd.columns[0:-1]:
    SF[f'Crowd_{col}']=subCrowd[col]

for year in range(2005,2021):
    if year==2020:
        year+=1
    subTotal1=SF_Jury[SF_Jury['Year']==year]
    subCrowd=SF_Crowd[SF_Crowd['Year']==year]
    for col in subCrowd.columns[0:-1]:
        subTotal1[f'Crowd_{col}']=subCrowd[col]
    SF=pd.concat([SF,subTotal1])
SF
```

```python
Final=Final_Jury[Final_Jury['Year']==1975]
Final
subCrowd=Final_Crowd[Final_Crowd['Year']==1975]
for col in subCrowd.columns[0:-1]:
    Final[f'Crowd_{col}']=subCrowd[col]

for year in range(1976,2021):
    if year==2020:
        year+=1
    subTotal1=Final_Jury[Final_Jury['Year']==year]
    subCrowd=Final_Crowd[Final_Crowd['Year']==year]
    for col in subCrowd.columns[0:-1]:
        subTotal1[f'Crowd_{col}']=subCrowd[col]
    Final=pd.concat([Final,subTotal1])
Final
```

```python
Final=pd.concat([Final,SF])
Final
```

```python
Final=Final.sort_values(by=['Year','Country'])
Final.reset_index(drop=True,inplace=True)
Final.drop(labels='Crowd_Unnamed: 0',axis=1,inplace=True)
```
